```python
full_data['Total'] = full_data['Total'].replace('[\$,]', '', regex=True).astype(float)
```

```python
cust_data = full_data.groupby('id')['Total'].sum().reset_index()
#grouping by users id
```

```python
cust_data['no_of_transactions'] = full_data.groupby('id')[['Date']].nunique().reset_index()['Date'].tolist()
cust_data
#getting the number of transactions made
```

```python
cust_data['no_of_items'] = full_data.groupby('id')[['Menu Item']].nunique().reset_index()['Menu Item'].tolist()
cust_data
#getting the number of unique items bought
```

```python
cust_data['loyalty_score'] = cust_data['no_of_items'] + cust_data['no_of_transactions'] + cust_data['Total']
#getting the loyalty score

```

```python
cust_data['scaled_loyalty_score'] = ((cust_data['loyalty_score'] -  cust_data['loyalty_score'].min()) / (cust_data['loyalty_score'].max() - cust_data['loyalty_score'].min())) * (10 - 1) + 1
#scaling loyalty score from 1-10
```

```python
cust_data['scaled_loyalty_score'] = cust_data['scaled_loyalty_score'] * 3
```

```python
cust_data[cust_data['scaled_loyalty_score'] >= 10] = 10
cust_data
```

```python

def recommend_items(cust_id, n=3):
  temp_df = vals.loc[cust_id].reset_index()
  #print(temp_df)
  max_rated_item_row = temp_df.loc[temp_df[cust_id].idxmax()]
  max_rated_item_row
  max_rated_item = max_rated_item_row['Menu Item']
  max_rated_item
  new_temp = item_item_sim_matrix.loc[max_rated_item].reset_index()
  new_temp = new_temp.rename(columns={max_rated_item: 'item_ratings'})
  new_temp
  top_n_items = new_temp.sort_values(by='item_ratings', ascending=False).head(n)
  return top_n_items['Menu Item'].tolist()
#recommending top n items
recommend_items('2520a9746b-ec13-4c6e-9848-c633b9848651na')
```
